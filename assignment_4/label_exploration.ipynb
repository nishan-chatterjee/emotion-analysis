```python
import pandas as pd
```

```python
ann1 = pd.read_csv("./dry_run_annotation_Isabelle.csv")
ann2 = pd.read_csv("./dry_run_annotation_Nishan.csv")
```

```python
ann1_trial = ann1[:10]
ann2_trial = ann2[:10]
ann1_main = ann1[10:]
ann2_main = ann2[10:]
```

```python
def simplify_annotations(ann1, ann2):
    # takes in two lists  of same length with words as entries
    a1 = []
    a2 = []
    for i in range(len(ann1)):
        if str(ann1.iloc[i]).lower() == str(ann2.iloc[i]).lower():
            a1.append(0)
            a2.append(0)
        elif str(ann1.iloc[i]).lower() in str(ann2.iloc[i]).lower().split():
            a1.append(0)
            a2.append(0)
        elif str(ann2.iloc[i]).lower() in str(ann1.iloc[i]).lower().split():
            a1.append(0)
            a2.append(0)
        else:
            a1.append(1)
            a2.append(2)
    return([a1,a2])
```

```python
from sklearn.metrics import cohen_kappa_score
```

```python
main_emo = simplify_annotations(ann1_main['emotion'], ann2_main['emotion'])
trial_emo = simplify_annotations(ann1_trial['emotion'], ann2_trial['emotion'])

trial_emo_agr = cohen_kappa_score(trial_emo[0],trial_emo[1])
main_emo_agr = cohen_kappa_score(main_emo[0],main_emo[1])
print("Trial 'emotion' agreement: ", round(trial_emo_agr,2))
print("Main 'emotion' agreement: ", round(main_emo_agr,2))
```

```python
main_exp = simplify_annotations(ann1_main['experiencer'], ann2_main['experiencer'])
trial_exp = simplify_annotations(ann1_trial['experiencer'], ann2_trial['experiencer'])

print(trial_exp)
trial_exp_agr = cohen_kappa_score(trial_exp[0],trial_exp[1])
main_exp_agr = cohen_kappa_score(main_exp[0],main_exp[1])
print("Trial 'experiencer' agreement: ", 1)
print("Main 'experiencer' agreement: ", round(main_exp_agr,2))
```

```python
main_cue = simplify_annotations(ann1_main['cue'], ann2_main['cue'])
trial_cue = simplify_annotations(ann1_trial['cue'], ann2_trial['cue'])

trial_cue_agr = cohen_kappa_score(trial_cue[0],trial_cue[1])
main_cue_agr = cohen_kappa_score(main_cue[0],main_cue[1])
print("Trial 'cue' agreement: ", round(trial_cue_agr,2))
print("Main 'cue' agreement: ", round(main_cue_agr,2))
```

```python
main_target = simplify_annotations(ann1_main['target'], ann2_main['target'])
trial_target = simplify_annotations(ann1_trial['target'], ann2_trial['target'])

trial_target_agr = cohen_kappa_score(trial_target[0],trial_target[1])
main_target_agr = cohen_kappa_score(main_target[0],main_target[1])
print("Trial 'target' agreement: ", round(trial_target_agr,2))
print("Main 'target' agreement: ", round(main_target_agr,2))
```

```python
main_stim = simplify_annotations(ann1_main['stimulus'], ann2_main['stimulus'])
trial_stim = simplify_annotations(ann1_trial['stimulus'], ann2_trial['stimulus'])

trial_stim_agr = cohen_kappa_score(trial_stim[0],trial_stim[1])
main_stim_agr = cohen_kappa_score(main_stim[0],main_stim[1])
print("Trial 'stimulus' agreement: ", round(trial_stim_agr,2))
print("Main 'stimulus' agreement: ", round(main_stim_agr,2))
```

```python
#have a look at how different our emotions are overall (trial and main together)

ann1_emotions = [str(x) for x in ann1['emotion'].tolist()]
ann2_emotions = [str(x) for x in ann2['emotion'].tolist()]

ann1_emo_count = {}
ann2_emo_count = {}
for l in ann1_emotions:
    if l in ann1_emo_count.keys():
        ann1_emo_count[l] += 1
    else:
        ann1_emo_count[l] = 1
        
for l in ann2_emotions:
    if l in ann2_emo_count.keys():
        ann2_emo_count[l] += 1
    else:
        ann2_emo_count[l] = 1
        
print(ann1_emo_count)
print(ann2_emo_count)
```

```python
#have a look at how different in length our cue, target and stimulus are (trial and main together)

ann1_cue = [str(x) for x in ann1['cue'].tolist()]
ann2_cue = [str(x) for x in ann2['cue'].tolist()]

ann1_target = [str(x) for x in ann1['target'].tolist()]
ann2_target = [str(x) for x in ann2['target'].tolist()]

ann1_stimulus = [str(x) for x in ann1['stimulus'].tolist()]
ann2_stimulus = [str(x) for x in ann2['stimulus'].tolist()]

def average_length(annotations): #takes in list of annotations
    lengths = []
    for an in annotations:
        lengths.append(len(an.split())) #how many words
    return round(sum(lengths)/len(lengths),2)

print("Annotator 1's 'cue' average length: ", average_length(ann1_cue))
print("Annotator 2's 'cue' average length: ", average_length(ann2_cue))
print("Annotator 1's 'target' average length: ", average_length(ann1_target))
print("Annotator 2's 'target' average length: ", average_length(ann2_target))
print("Annotator 1's 'stimulus' average length: ", average_length(ann1_stimulus))
print("Annotator 2's 'stimulus' average length: ", average_length(ann2_stimulus))
```
